```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_highest-grossing_films"
response = requests.get(url)
status = BeautifulSoup(response.content, "html.parser")
output =status.find("table", class_="wikitable")
description = status.find("caption").text.replace("\n","").replace("[12]","")
description
```

```python
headers = []
for x in output.find_all("tr"):
    for i in x.find_all("th"):
        tex_t = i.text.replace("\n","").strip()
        headers.append(tex_t)
df = pd.DataFrame(columns=headers)
df
```

```python
drop_columns = df.columns[6:57]
drops = df.drop(columns=drop_columns)
drops
df = pd.DataFrame(columns=drops.columns)
df
```

```python
tbody = output.find("tbody")
data_list = []

for tr in tbody.find_all("tr"):
    row_data = []
    for td in tr.find_all("td"):
        row_data.append(td.text.replace("\n", ""))
    for td in tr.find_all("td", class_="data-sort-value"):
        row_data.append(td.text)
    for td in tr.find_all("i"):
        row_data.append(td.text)
    data_list.append(row_data)
data_list
```

```python
data_list.pop(0)
data_list
```

```python
newframe = pd.DataFrame(data_list, columns=drops.columns)
newframe
```
