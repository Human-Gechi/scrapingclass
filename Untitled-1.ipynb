```python
import requests
from bs4 import BeautifulSoup

url = "https://www.jumia.com.ng/computing/"
response = requests.get(url)
output = BeautifulSoup(response.content, "html.parser")

names =[]

data = output.find_all("div", class_="info")
for info in data:
        name = info.find("h3", class_="name")
        names.append(name.text)
names
```

```python
new_prices= []
old_prices = []
for info in data:
    discount_price = info.find("div", class_="prc").text
    old_price = info.find("div", class_="s-prc-w")
    actual_price = old_price.text if old_price else "No old price"
    new_prices.append(discount_price)
    old_prices.append(actual_price)

old_prices
new_prices
```

```python
ratings = []
for info in data:
    rating = info.find("div", class_="rev").text
    ratings.append(rating)
ratings
```

```python
import pandas as pd
products_df = pd.DataFrame({
    "names":names,\
    "oldprices":old_prices,
    "newprices":new_prices,
    "rating":ratings
})
products_df
```

```python
products_df[["rating_stars","total_rating"]] = products_df["rating"].str.split(" ",expand=True)
```

```python
products_df = products_df.drop(columns=[("rating_stars","total_rating")])
products_df
```
