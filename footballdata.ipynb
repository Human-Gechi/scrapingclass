```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.worldfootball.net/players_list/eng-premier-league-2024-2025/nach-name/1/"
response = requests.get(url=url)
soup = BeautifulSoup(response.content ,"html.parser")
output = soup.find("table", class_ = "standard_tabelle")
output
```

```python
headers = []
for x in output.find_all("tr"):
    for i in x.find_all("th"):
        tex_t = i.text
        headers.append(tex_t)
headers
df = pd.DataFrame(columns=headers)
df
```

```python

data = []
for j in output.find_all("tr")[1:]:
        row_data = j.find_all("td", class_="hell") 
        if not row_data:
            row_data= j.find_all("td",class_ = "dunkel")
            row_values = [row.text for row in row_data]
            data.append(row_values)
new_data = pd.DataFrame(data, columns=headers)
new_data
```
